```python
import sqlite3
import csv
```

## Create and connect to database

```python
con = sqlite3.connect("air_pollution_database.db")
cur = con.cursor()
```

## Create table for site: AJ Jacobs

```python
cur.execute('''CREATE TABLE AJ_Jacobs (time int, day int, month int, year int, 
  SO2 float, NO2 float, NO float, NOx float, CO float, PM2_5 float, PM10 float, 
  Amb_Wspeed float, Amb_WDirection float)''')

con.commit()

data_file = open("AJ_Jacobs.csv")
rows = csv.reader(data_file)
cur.executemany("INSERT INTO AJ_Jacobs VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", rows)

con.commit()

data_file = open("AJ_Jacobs-2.csv")
rows = csv.reader(data_file)
cur.executemany("INSERT INTO AJ_Jacobs VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", rows)

con.commit()
```
